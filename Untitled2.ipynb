```python
import tensorflow as tf

#downloading the dataset
data = tf.keras.datasets.mnist
(trainX2D, trainy), (testX2D, testy)=data.load_data()
print(type(trainX2D))
print(trainX2D.shape)
```

```python
#trainX contains the 1D data inputs
trainX = trainX2D.reshape(trainX2D.shape[0],784)
testX = testX2D.reshape(testX2D.shape[0],784)
trainX = trainX.astype('float32')
testX = testX.astype('float32')
trainX /= 255
testX /= 255
#Now we'll change the format of the output
#from cateogrical data to binary
#trainy contains categorical data
#train_y is its converted form
train_y = tf.keras.utils.to_categorical(trainy,10)
test_y = tf.keras.utils.to_categorical(testy,10)
```

```python
#these are the place holders for input images and its labels
inputim = tf.placeholder(tf.float32, shape=[None, 784])
targetlb = tf.placeholder(tf.float32, shape=[None, 10])


```

```python

#number of hidden nodes
hidden_nodes = 400
#weights between layer 1 and 2
w12= tf.Variable(tf.truncated_normal([784, hidden_nodes]))
#biases for layer one set to 0
in_biases = tf.Variable(tf.zeros([hidden_nodes]))


#weights between layer 2 and 3
w23= tf.Variable(tf.truncated_normal([hidden_nodes,10]))
hidd_biases = tf.Variable(tf.zeros([10]))
```

```python


#creating the flow of tensors basicallly
input_layer = tf.matmul(inputim, w12)
hidden_layer = tf.nn.relu(input_layer + in_biases)
digit_weights = tf.matmul(hidden_layer, w23) + hidd_biases

```

```python
#This is the cost function that I am using
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=digit_weights, labels=targetlb))
#And we are using gradient Descent optimizer
optimizer = tf.train.GradientDescentOptimizer(0.5).minimize(cross_entropy)

```

```python
correct_prediction = tf.equal(tf.argmax(digit_weights,1), tf.argmax(targetlb,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
```
